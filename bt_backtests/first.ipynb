```python
import bt
import matplotlib as plt
import pandas as pd

%matplotlib inline

plt.rcParams['figure.dpi'] = 500

SYMBOLS = ['AAPL', 'MSFT', 'C', 'GS', 'GE']
data = bt.get(','.join(SYMBOLS), start='2010-01-01')
```

```python
sma = data.rolling(50).mean()
```

```python
plot = bt.merge(data, sma).plot(figsize=(15, 5))
```
